# Three-level toy-model system

In this project we will consider a toy-model which main goal is to show the difference between the fast quasidiabatic method and the optimization of the excess work. The Hamiltonian of the system is given by 
\begin{equation}
H(\lambda(t)) = \frac{\hbar}{2}\begin{pmatrix}
0 & \Omega_p & 0 \\
\Omega_p & 2 \Delta & \Omega_s \\
0 & \Omega_s & \delta 
\end{pmatrix} = \begin{pmatrix}
0 & a & 0 \\
a & \lambda(t) & a \\
0 & a & 0 
\end{pmatrix} 
\end{equation}

where $a$ is constant and $b$ the control parameter. Its eigenvalues are given by
\begin{equation}
\begin{split}
    E_{+}(\lambda(t)) & = \frac{\left( \lambda(t) + \sqrt{\lambda(t)^2 + 8a^2}\right) }{2}   \\
    E_{-}(\lambda(t)) &= \frac{\left( \lambda(t) - \sqrt{\lambda(t)^2 + 8a^2}\right) }{2}  \\ 
    E_{0}(\lambda(t)) &= 0 \\
\end{split}
\end{equation}
while its eigenstates
\begin{equation}
\begin{split}
 | n_{+}(t) \rangle &= \frac{\sqrt{2}}{2} \sin \phi(t) | 1 \rangle + \cos \phi(t) | 2 \rangle  + \frac{\sqrt{2}}{2}  \sin \phi(t) | 3 \rangle  \\
 | n_{-}(t) \rangle &= \frac{\sqrt{2}}{2}\cos \phi(t) | 1 \rangle - \sin \phi(t) | 2 \rangle + \frac{\sqrt{2}}{2}\cos \phi(t) | 3 \rangle  \\
 | n_{0}(t) \rangle &=  \frac{\sqrt{2}}{2}\left( | 1 \rangle  -  | 3 \rangle \right)
\end{split}
\end{equation}
where
\begin{equation}
\tan 2 \phi(t) = \frac{2 \sqrt{2}a}{\lambda(t)}
\end{equation}

```python
import math

# Positive eigenvalue
def Eplus(a,b):
    return 0.5*(b + math.sqrt(8*a**2 + b**2)) 
# Zero eigenvalue
def Ezero(a,b):
    return 0.0
#Negative eigenvalue
def Eminus(a,b):
    return 0.5*(b - math.sqrt(8*a**2 + b**2)) 
```

To verify the behavior of the eigenvalues, lets consider $a = 0.1$ and a linear protocol for $b(t) = t$. Then,

## The counterdiabatic Hamiltonian

\begin{equation}
\begin{split}
 | n_{+}(t) \rangle &= \frac{\sqrt{2}}{2} \sin \phi(t) | 1 \rangle + \cos \phi(t) | 2 \rangle  + \frac{\sqrt{2}}{2}  \sin \phi(t) | 3 \rangle  \\
 | n_{-}(t) \rangle &= \frac{\sqrt{2}}{2}\cos \phi(t) | 1 \rangle - \sin \phi(t) | 2 \rangle + \frac{\sqrt{2}}{2}\cos \phi(t) | 3 \rangle  \\
 | n_{0}(t) \rangle &=  \frac{\sqrt{2}}{2}\left( | 1 \rangle  -  | 3 \rangle \right)
\end{split}
\end{equation}

The counterdiabatic Hamiltonian can be written as
\begin{equation}
H_{1}(t) = i \hbar \sum_{m,n,m \neq n} \langle m(t) | \frac{d}{dt} |n (t) \rangle|m(t)\rangle \langle n(t)|.
\end{equation}
Therefore, for the toy-model 
\begin{equation}
\begin{split}
 \frac{d}{dt}| n_{+}(t) \rangle &= \dot{\phi}(t)\left[\frac{\sqrt{2}}{2} \cos \phi(t) | 1 \rangle - \sin \phi(t) | 2 \rangle  + \frac{\sqrt{2}}{2}  \cos \phi(t) | 3 \rangle\right]  \\
 \frac{d}{dt}| n_{-}(t) \rangle &= \dot{\phi}(t)\left[-\frac{\sqrt{2}}{2}\sin \phi(t) | 1 \rangle - \cos \phi(t) | 2 \rangle - \frac{\sqrt{2}}{2}\sin \phi(t) | 3 \rangle \right] \\
 \frac{d}{dt}| n_{0}(t) \rangle &=  0
\end{split}
\end{equation}
